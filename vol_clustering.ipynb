```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
```

```python
paths = torch.load('./data/long_paths_p3_q3.pt')
paths.shape
```

```python
log_returns = paths.log().diff(dim=1)
```

```python
size = 10
row = 2
indices = np.random.randint(0, log_returns.shape[0], size=size)
fig, ax = plt.subplots(size // row, row, figsize=(row*8, size//row*6))
for i in range(size):
    ax[i//row, i%row].plot(log_returns[indices[i]], label='CSigWGAN Generator');
```
